```python
from main import create_messi_data_transitions_ledger

messi_transitions_ledger = create_messi_data_transitions_ledger()
print(len(messi_transitions_ledger))
```

```python
from main import build_Q_matrix


# 1. Erase the floating-point '.0' ghosts so the clones merge back together
messi_transitions_ledger["state_id"] = messi_transitions_ledger["state_id"].str.replace(
    ".0", "", regex=False
)
messi_transitions_ledger["next_state_id"] = messi_transitions_ledger[
    "next_state_id"
].str.replace(".0", "", regex=False)

# 2. Filter out the out-of-bounds 'nan' states
clean_ledger = messi_transitions_ledger[
    ~messi_transitions_ledger["state_id"].str.contains("nan")
]
clean_ledger = clean_ledger[~clean_ledger["next_state_id"].str.contains("nan")]

# 3. Rebuild the Matrix
Q_matrix = build_Q_matrix(clean_ledger)

filename = "./datasets/messi_data_Q_matrix"
Q_matrix.to_csv(filename, index=False)
print(f"Saved to {filename}")

print(f"Purged Ghost States. New Q-Matrix Shape: {Q_matrix.shape}")
```
